```python
# Import the necessary library
import pandas as pd
```

```python
# Define the file path to your original CSV
csv_file_path = "Source_Data/energy-pop-exposure-nuclear-plants-locations_plants.csv"
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)
```

```python
# Display the first few rows to inspect the data (optional)
print("Initial Data:")
display(df.head())
```

```python
# Filter the DataFrame to only keep rows where a column contains 'United States of America'
filtered_df = df[df['Country'] == 'UNITED STATES OF AMERICA']
```

```python
# Display the filtered data (optional)
display(filtered_df.head())
```

```python
# Drop columns 8 through 61 (index-based)
df_dropped = filtered_df.drop(filtered_df.columns[7:62], axis=1)

```

```python
# Display the updated DataFrame (optional)
print("Data after dropping columns 8 through 61:")
display(df_dropped.head())
```

```python
# Save the filtered DataFrame to a new CSV file
cleaned_csv_path = "Source_Data/cleaned_nuclear_data.csv"  # Define the path and name of the new CSV
df_dropped.to_csv(cleaned_csv_path, index=False)
```

```python
print(f"Filtered CSV has been saved as '{cleaned_csv_path}'.")
```
